```python
from gensim.models import Word2Vec
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Masters_research/YouTube_videos.csv')
df.dropna(inplace=True)
df2=pd.read_csv('/content/drive/MyDrive/Masters_research/processed_ads')
df2.dropna(inplace=True)
video_titles=df['Title'].astype(str)
advertisement_titles=df2['title'].astype(str)

# Step 2: Train Word2Vec embeddings
combined_corpus = video_titles.tolist() + advertisement_titles.tolist()  # Concatenate your datasets
model = Word2Vec(combined_corpus, vector_size=100, window=5, min_count=1)  # Train Word2Vec model

# Step 3: Generate representations
# Step 3: Generate representations
video_title_representations = []
for video_title in video_titles:
    words = video_title.split()  # Tokenize into individual words
    video_title_representation = np.mean([model.wv[word] for word in words if word in model.wv], axis=0)  # Average the word embeddings
    video_title_representations.append(video_title_representation)

advertisement_title_representations = []
for ad_title in advertisement_titles:
    words = ad_title.split()  # Tokenize into individual words
    ad_title_representation = np.mean([model.wv[word] for word in words if word in model.wv], axis=0)  # Average the word embeddings
    advertisement_title_representations.append(ad_title_representation)

# Step 4: Perform recommendation for each video title
top_k = 5  # Define the number of recommendations you want to display

for video_title, video_representation in zip(video_titles, video_title_representations):
    similarities = []
    for ad_title, ad_representation in zip(advertisement_titles, advertisement_title_representations):
        if np.isnan(ad_representation).any():
            continue  # Skip empty representations

        similarity = np.dot(video_representation, ad_representation) / (
            np.linalg.norm(video_representation) * np.linalg.norm(ad_representation)
        )
        similarities.append(similarity)

    recommended_indices = np.argsort(similarities)[::-1]  # Sort in descending order
    recommended_advertisement_titles = [advertisement_titles[i] for i in recommended_indices]

    # Print the top recommended advertisement titles
    print(f"Recommended Advertisement Titles for '{video_title}':")
    for title in recommended_advertisement_titles[:top_k]:
        print(title)
    print()  # Add a new line between recommendations for different video titles
```
